Tokenizing a summarized Completion

```python
from transformers import AutoTokenizer,pipeline


classifier = pipeline("summarization",model = "sshleifer/distilbart-cnn-12-6")
text=classifier("Paris is the capital and most populous city of France, with an estimated population of 2,175,601 residents as of 2018, in an area of more than 105 square kilometres (41 square miles). The City of Paris is the centre and seat of government of the region and province of Île-de-France, or Paris Region, which has an estimated population of 12,174,880, or about 18 percent of the population of France as of 2017.",
           max_length=30,  # Set the maximum length of the summary
    min_length=20     )

# Load a pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

summary_text = text[0]['summary_text']  # Assuming the summary is in the first element of the list

# Tokenize the summary text
tokens = tokenizer.tokenize(summary_text)
print("Tokens:", tokens)
```

```python
# Convert tokens to input IDs
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)
```

```python
# Encode the text (tokenization + converting to input IDs)
encoded_input = tokenizer(summary_text)
print("Encoded Input:", encoded_input)
```

```python
# Decode the text
decoded_output = tokenizer.decode(input_ids)
print("Decode Output: ", decoded_output)
```

```python
from datasets import load_dataset
dataset = load_dataset('imdb')
```

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['text'], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python

tokenized_datasets
```

```python
tokenized_datasets["train"][0]
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # Output directory
    eval_strategy ="epoch",     # Evaluate every epoch
    learning_rate=2e-5,              # Learning rate
    per_device_train_batch_size=16,  # Batch size for training
    per_device_eval_batch_size=16,   # Batch size for evaluation
    num_train_epochs=3,              # Number of training epochs
    weight_decay=0.01,               # Strength of weight decay
    run_name='./my_imdb_training',   # changed the run_name_ the run_name for your training run is currently the same as the output_dir specified in your TrainingArguments.
)
training_args
```

```python
from transformers import AutoModelForSequenceClassification, Trainer

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test']
)

```

```python

# Train the model
trainer.train()
```

```python
# Save the model
model.save_pretrained('./fine-tuned-model')
tokenizer.save_pretrained('./fine-tuned-tokenizer')
```
